```python
import pandas as pd
from apyori import apriori
```

```python
data=pd.read_csv(r'C:\Users\user\Music\ml udemy\Untitled Folder\Unsupervised Learning\Apriori\groceries.csv')
```

```python
data
```

```python
data1=data.drop(['Item(s)'],axis=1)
```

```python
data1
```

```python
purchase=[]
for i in range(0,9835):
    purchase.append([str(data1.values[i,j]) for j in range(0,32)])
```

```python
purchase
```

```python
rules=apriori(transactions=purchase,min_support=0.00213,min_confidence=0.15,min_lift=2,min_length=2,max_length=2)
```

```python
results=list(rules)
```

```python
results
```

```python
def inspect(results):
    lhs        =[tuple(result[2][0][0])[0] for result in results]
    rhs        =[tuple(result[2][0][1])[0] for result in results]
    support    =[result[1] for result in results]
    confidence =[result[2][0][2] for result in results]
    lift       =[result[2][0][3] for result in results]
    return list(zip(lhs,rhs,support,confidence,lift))
resultinDataFrame=pd.DataFrame(inspect(results),columns=['Left Hand Side','Right Hand Side','Support','Confidence','Lift'])
    
```

```python
resultinDataFrame.sort_values('Lift',ascending=False)
```

```python
resultinDataFrame.nlargest(n=11,columns='Lift')
```
